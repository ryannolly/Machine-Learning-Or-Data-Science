Sebagai contoh kita memiliki data sebagai berikut :

$x$ | 0 | 20 | 40 | 60 | 100 | 120
--- | --- | --- | --- | --- | --- | --- 
$y$ | 26.0 | 48.6 | 60 | 30 | 100 | -1

Dan kita akan mencari grafik hasil dari interpolasi dengan data tersebut

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0 , 20, 40, 60, 100, 120]
y = [26.0, 48.6, 60, 30, 100, -1]

m = len(x)
n = m - 1 #Derajat dari fungsi

xplt = np.linspace(x[0], x[-1], num = 100)
yplt = []

for i in range(len(xplt)) :
  yp = 0
  for j in range(n + 1) : 
    p = 1
    for k in range(n + 1) :
      if k != j :
        p *= (xplt[i] - x[k])/(x[j] - x[k])
    yp += y[j] * p
  yplt.append(yp)

# print(xplt)
# print(yplt)

plt.figure(figsize=(20, 15))
plt.plot(x, y, 'bo')
plt.plot(xplt, yplt, 'r-')
```
