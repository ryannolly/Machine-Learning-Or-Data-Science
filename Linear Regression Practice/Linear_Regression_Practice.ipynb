Berikut ini data pengalaman kerja dan penjualan

$X = $ Pengalaman Kerja (Tahun)

$Y = $ Omset Penjualan (Ribuan)

$X$ | 2 | 3 | 2 | 5 | 6 | 1 | 4 | 1 |
--- | --- | --- | --- | --- | --- | --- | --- | --- |
$Y$ | 5 | 8 | 8 | 7 | 11 | 3 | 10 | 4 |

* Tentukan Nilai a dan b! (Dari Persamaan $Y = a + bx$)
* Buatkan Persamaan Regresinya
* Berapa omzet penjualan dari seorang karyawan yang pengalaman kerjanya 3.5 Tahun?



```python
# Import module untuk ngeplot grafik nanti
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x_values = [2, 3, 2, 5, 6, 1, 4, 1]
y_values = [5, 8, 8, 7, 11, 3, 10, 4]

plt.scatter(x_values, y_values)
plt.xlabel("Pengalaman Kerja (Tahun)")
plt.ylabel("Omset Penjualan (Ribuan)")
plt.show()
```

Setelah itu, kita akan melakukan regresi dengan mencari nilai a dan b dari fungsi $Y = a + bx$ dengan menggunakan Least Square Method. Adapun rumusnya adalah sebagai berikut : 

image.png

```python
# Mencari SigmaXY
sigma_xy = 0
for i in range(len(x_values)) :
  sigma_xy = sigma_xy + (x_values[i] * y_values[i])

# Mencari Sigma X
sigma_x = 0
for i in x_values : 
  sigma_x = sigma_x + i

# Mencari Sigma Y
sigma_y = 0
for i in y_values : 
  sigma_y = sigma_y + i

# Mencari Sigma X kuadrat
sigma_x_squared = 0
for i in x_values : 
  sigma_x_squared = sigma_x_squared + (i ** 2)

print(sigma_xy, sigma_x, sigma_y, sigma_x_squared) 
```

Setelah itu didapat nilai b sebagai berikut : 

```python
nilai_b = ((len(x_values) * sigma_xy) - (sigma_x * sigma_y)) / ((len(x_values) * sigma_x_squared) - (sigma_x ** 2))
print("Nilai B nya adalah : ", nilai_b)
```

Setelah itu kita akan mencari a dengan mencari x bar dan y bar terlebih dahulu

```python
x_bar = 0
for i in x_values : 
  x_bar = x_bar + i

x_bar = x_bar / len(x_values)
y_bar = 0
for i in y_values : 
  y_bar = y_bar + i
y_bar = y_bar / len(y_values)
print(x_bar, y_bar)
```

Setelah itu, kita akan menggunakan rumus berikut untuk mencari a 

$a = Y - 1.25X$

```python
nilai_a = y_bar - (nilai_b * x_bar)
print(nilai_a)
```

Sehingga fungsi regresinya adalah : 
$Y = 3.25 + 1.25X$

Dan apabila diplot dengan datanya akan menjadi seperti berikut : 

```python
def function_y(x) : 
  return 3.25 + (1.25 * x)

nilai_x_predicted = []
nilai_y_predicted = []

for i in range(10) : 
  nilai_x_predicted.append(i)
  nilai_y_predicted.append(function_y(i))

print(nilai_x_predicted)
print(nilai_y_predicted)

plt.title("Hasil regresi dengan data asli")
plt.plot(nilai_x_predicted, nilai_y_predicted, marker="o")
plt.scatter(x_values, y_values, color='y')
plt.xlabel("Pengalaman Kerja (Tahun)")
plt.ylabel("Omset Penjualan (Ribuan)")
plt.show()
```

Hasil Omset penjualan dalam 3.5 Tahun adalah :

```python
print("Hasil penjualan 3.5 Tahun : ", function_y(3.5))
```
